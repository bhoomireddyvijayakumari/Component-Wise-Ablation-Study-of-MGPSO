```python
"""
mgpso_only_pm.py  —  ONLY_PM
-----------------------
Gbest + random archive guide + polynomial mutation  |  ONLY PM added
USE_RING=False  USE_TS=False  USE_PM=True

Speed: particle loop fully vectorised with NumPy (~9x faster than loop).
30 runs x 14 benchmarks x sensitivity sweeps runs in one terminal.

PART 1 : Main experiment  (default params, 30 runs x all 14 benchmarks)
         Pareto fronts shown inline (Jupyter) and saved to disk.
PART 2 : Sensitivity sweep (active components only) [DISABLED]
         ts_size : [2, 3, 5, 7, 10]   (USE_TS=False)
         eta     : [5, 10, 20, 30, 50]  (USE_PM=True)
         k       : [1, 2, 3, 4, 5]    (USE_RING=False)

Usage:
  python mgpso_only_pm.py          # all 14 benchmarks
  python mgpso_only_pm.py zdt1     # one benchmark only
  %run   mgpso_only_pm.py          # Jupyter
"""

import os, sys
import numpy as np
import pandas as pd
import matplotlib

_in_jupyter = False
try:
    __file__
except NameError:
    _in_jupyter = True

if not _in_jupyter:
    matplotlib.use("Agg")

import matplotlib.pyplot as plt
from tqdm import tqdm
from pymoo.core.population import Population
from pymoo.problems import get_problem
from pymoo.indicators.igd import IGD
from pymoo.indicators.hv import HV
from pymoo.algorithms.moo.nsga2 import RankAndCrowdingSurvival

plt.rcParams.update({
    "font.family": "serif", "font.size": 11,
    "axes.labelsize": 13, "axes.titlesize": 13,
    "legend.fontsize": 10, "xtick.labelsize": 10, "ytick.labelsize": 10,
    "figure.dpi": 100, "savefig.dpi": 200,
    "savefig.bbox": "tight", "axes.grid": True, "grid.alpha": 0.3,
})

# ── Constants ─────────────────────────────────────────────────────────
N_PARTICLES  = 150
ARCHIVE_SIZE = 300
N_RUNS       = 30
W            = 0.7298
C1 = C2 = C3 = 1.49618
LAMBD        = 0.5
DEFAULT_TS   = 5
DEFAULT_ETA  = 20
DEFAULT_K    = 3
TS_GRID      = [2, 3, 5, 7, 10]
ETA_GRID     = [5, 10, 20, 30, 50]
K_GRID       = [1, 2, 3, 4, 5]

VARIANT_NAME = "ONLY_PM"
USE_RING     = False
USE_TS       = False
USE_PM       = True
COLOR        = "#9C27B0"

ALL_BENCHMARKS = [
    "zdt1", "zdt2", "zdt3", "zdt4", "zdt6",
    "wfg1", "wfg2", "wfg3", "wfg4",
    "wfg5", "wfg6", "wfg7", "wfg8", "wfg9",
]

# ── Utilities ─────────────────────────────────────────────────────────

def polynomial_mutation(X, xl, xu, eta):
    n, d     = X.shape
    prob     = 1.0 / d
    mut_mask = np.random.random((n, d)) < prob
    delta    = np.where((xu - xl) < 1e-10, 1e-10, xu - xl)
    u        = np.random.random((n, d))
    delta1   = np.clip((X - xl) / delta, 0, 1)
    delta2   = np.clip((xu - X) / delta, 0, 1)
    pow_val  = 1.0 / (eta + 1.0)
    mask_low = u <= 0.5
    val_low  = 2.0*u + (1.0-2.0*u)*np.power(np.clip(1.0-delta1, 0, None), eta+1.0)
    dq_low   = np.power(np.clip(val_low,  0, None), pow_val) - 1.0
    val_high = 2.0*(1.0-u) + 2.0*(u-0.5)*np.power(np.clip(1.0-delta2, 0, None), eta+1.0)
    dq_high  = 1.0 - np.power(np.clip(val_high, 0, None), pow_val)
    deltaq   = np.where(mask_low, dq_low, dq_high)
    return np.clip(X + deltaq * delta * mut_mask, xl, xu)

def update_archive(problem, archive, new_pop):
    combined = (new_pop if (archive is None or len(archive) == 0)
                else Population.merge(archive, new_pop))
    survival = RankAndCrowdingSurvival()
    return survival.do(problem, combined, n_survive=min(len(combined), ARCHIVE_SIZE))

def tournament_selection(archive, ts_size):
    """Vectorised: draw ts_size candidates, return highest-crowding-distance one."""
    if len(archive) == 0: return None
    idx   = np.random.choice(len(archive), min(ts_size, len(archive)), replace=False)
    cands = archive[idx]
    cds   = cands.get("crowding")
    if cds is not None:
        return cands[np.argmax(np.nan_to_num(cds, nan=0.0, posinf=1e6))].x
    return cands[np.random.randint(len(cands))].x

def random_selection(archive):
    if len(archive) == 0: return None
    return archive[np.random.randint(len(archive))].x

# ── Vectorised pbest update ───────────────────────────────────────────

def update_pbest(pbest_X, pbest_F, X, F):
    """Update personal bests for all particles at once with NumPy."""
    # Dominance checks — fully vectorised
    new_dom_old = np.all(F <= pbest_F, axis=1) & np.any(F < pbest_F, axis=1)
    old_dom_new = np.all(pbest_F <= F, axis=1) & np.any(pbest_F < F, axis=1)
    non_dom     = ~new_dom_old & ~old_dom_new
    rand_accept = np.random.random(len(X)) < 0.5
    update_mask = new_dom_old | (non_dom & rand_accept)
    pbest_X[update_mask] = X[update_mask].copy()
    pbest_F[update_mask] = F[update_mask].copy()
    return pbest_X, pbest_F

# ── Core optimisation loop (vectorised) ───────────────────────────────

def run_mgpso(problem_obj, n_gen, seed=0,
              ts_size=DEFAULT_TS, eta=DEFAULT_ETA, k=DEFAULT_K):
    np.random.seed(seed)
    xl, xu = problem_obj.bounds()
    vmax   = 0.2 * (xu - xl)
    N, D   = N_PARTICLES, problem_obj.n_var

    X = np.random.uniform(xl, xu, (N, D))
    V = np.random.uniform(-0.1*(xu-xl), 0.1*(xu-xl), (N, D))
    F       = problem_obj.evaluate(X)
    
    # FIX: Explicit keyword arguments for modern pymoo API compatibility
    pop     = Population.new(X=X, F=F)
    
    pbest_X = np.copy(X)
    pbest_F = np.copy(F)
    archive = update_archive(problem_obj, Population(), pop)

    # Precompute ring neighbourhood indices once (only used if USE_RING)

    for gen in range(n_gen):

        gbest   = np.argmin(pbest_F.sum(axis=1))
        Y_HAT   = np.tile(pbest_X[gbest], (N, 1))     # (N, D)

        a_guide = random_selection(archive)
        A_HAT   = np.tile(a_guide, (N, 1)) if a_guide is not None else Y_HAT

        # Vectorised velocity & position update
        R1 = np.random.random((N, D))
        R2 = np.random.random((N, D))
        R3 = np.random.random((N, D))
        V  = (W * V
              + C1 * R1               * (pbest_X - X)
              + LAMBD         * C2 * R2 * (Y_HAT   - X)
              + (1 - LAMBD) * C3 * R3 * (A_HAT   - X))
        V  = np.clip(V, -vmax, vmax)
        X  = np.clip(X + V, xl, xu)

        X = polynomial_mutation(X, xl, xu, eta)

        F   = problem_obj.evaluate(X)
        
        # FIX: Explicit keyword arguments for modern pymoo API compatibility
        pop = Population.new(X=X, F=F)
        
        pbest_X, pbest_F = update_pbest(pbest_X, pbest_F, X, F)
        archive = update_archive(problem_obj, archive, pop)

    return archive

# ── Benchmark config ──────────────────────────────────────────────────

def get_bench(b_name):
    is_wfg    = b_name.lower().startswith("wfg")
    problem   = (get_problem(b_name, n_var=24, n_obj=3, k=4)
                 if is_wfg else get_problem(b_name))
    n_gen     = 400 if is_wfg else (500 if b_name == "zdt4" else 300)
    pf        = problem.pareto_front()
    ref_point = (np.array([3.0, 5.0, 7.0]) if is_wfg
                 else np.ones(problem.n_obj) * 2.5)
    return problem, n_gen, pf, ref_point

# ── Trial loop ────────────────────────────────────────────────────────

def trial_loop(problem, n_gen, pf, ref_point, run_kwargs, label):
    rows = []
    best_igd, best_F = np.inf, None
    pbar = tqdm(range(N_RUNS), desc=f"  {label}", unit="run", ncols=80, leave=True)
    for seed in pbar:
        arc = run_mgpso(problem, n_gen, seed=seed, **run_kwargs)
        F   = arc.get("F")
        igd = IGD(pf).do(F) if pf is not None else np.nan
        hv  = HV(ref_point=ref_point).do(F)
        rows.append({"seed": seed, "IGD": igd, "HV": hv})
        if igd < best_igd:
            best_igd, best_F = igd, F.copy()
        pbar.set_postfix(IGD=f"{igd:.4f}", best=f"{best_igd:.4f}")
    return pd.DataFrame(rows), best_F

# ── Pareto plot ───────────────────────────────────────────────────────

def plot_pareto(pf, F, title, path):
    n_obj = F.shape[1]
    if n_obj == 2:
        fig, ax = plt.subplots(figsize=(7, 5))
        if pf is not None:
            s = pf[pf[:,0].argsort()]
            ax.plot(s[:,0], s[:,1], color="black", lw=2.0,
                    label="True Pareto Front", alpha=0.85, zorder=5)
        ax.scatter(F[:,0], F[:,1], facecolor="none", edgecolor=COLOR,
                   s=45, linewidths=1.3, label=VARIANT_NAME, alpha=0.9, zorder=10)
        ax.set_xlabel(r"$f_1$"); ax.set_ylabel(r"$f_2$")
    else:
        fig = plt.figure(figsize=(9, 7))
        ax  = fig.add_subplot(111, projection="3d")
        if pf is not None:
            ax.scatter(pf[:,0], pf[:,1], pf[:,2], c="gray", s=6,
                       label="True PF", alpha=0.15, zorder=1)
        ax.scatter(F[:,0], F[:,1], F[:,2], facecolor="none", edgecolor=COLOR,
                   s=28, linewidths=0.9, label=VARIANT_NAME, alpha=0.85, zorder=10)
        ax.set_xlabel(r"$f_1$", labelpad=7)
        ax.set_ylabel(r"$f_2$", labelpad=7)
        ax.set_zlabel(r"$f_3$", labelpad=7)
        ax.view_init(elev=25, azim=45)
    ax.set_title(title, pad=10)
    ax.legend(loc="best", framealpha=0.9)
    plt.tight_layout()
    plt.savefig(path, dpi=200, bbox_inches="tight")
    plt.show()
    plt.close()

# ── Summary helpers ───────────────────────────────────────────────────

def save_summary(rows, out_dir, prefix):
    summary = pd.DataFrame(rows)
    summary.to_csv(f"{out_dir}/{prefix}_summary.csv", index=False)
    for metric in ("IGD", "HV"):
        pt = summary[["Function", f"{metric}_Mean", f"{metric}_Std"]].copy()
        pt[metric] = pt.apply(
            lambda r, m=metric: f"{r[f'{m}_Mean']:.4e} +/- {r[f'{m}_Std']:.2e}", axis=1)
        pt[["Function", metric]].set_index("Function").to_csv(
            f"{out_dir}/{prefix}_pivot_{metric}.csv")
    return summary

def plot_line(sweep_vals, results, metric, param_label, title, path, benchmarks):
    cmap   = plt.cm.tab20
    colors = [cmap(i / max(len(benchmarks)-1, 1)) for i in range(len(benchmarks))]
    fig, ax = plt.subplots(figsize=(9, 5))
    all_y = []
    for ci, b in enumerate(benchmarks):
        ys = [results.get((b, v), {}).get(f"{metric}_Mean", np.nan) for v in sweep_vals]
        ax.plot(sweep_vals, ys, marker="o", lw=1.3, color=colors[ci],
                alpha=0.65, label=b.upper())
        all_y.append(ys)
    avg = np.nanmean(all_y, axis=0)
    ax.plot(sweep_vals, avg, marker="D", lw=2.5, color="black",
            label="Average", zorder=10)
    ax.set_xlabel(param_label); ax.set_ylabel(f"Mean {metric}")
    ax.set_title(title); ax.set_xticks(sweep_vals)
    ax.legend(loc="best", framealpha=0.9, fontsize=8, ncol=2)
    plt.tight_layout()
    plt.savefig(path, dpi=200, bbox_inches="tight")
    plt.show()
    plt.close()

def plot_heatmap(sweep_vals, results, metric, param_label, title, path, benchmarks):
    data = np.full((len(benchmarks), len(sweep_vals)), np.nan)
    for bi, b in enumerate(benchmarks):
        for vi, v in enumerate(sweep_vals):
            data[bi, vi] = results.get((b, v), {}).get(f"{metric}_Mean", np.nan)
    fig, ax = plt.subplots(figsize=(max(6, len(sweep_vals)*1.5),
                                    max(4, len(benchmarks)*0.5)))
    cmap = "RdYlGn_r" if metric == "IGD" else "RdYlGn"
    im = ax.imshow(data, aspect="auto", cmap=cmap)
    plt.colorbar(im, ax=ax, label=f"Mean {metric}")
    ax.set_xticks(range(len(sweep_vals)))
    ax.set_xticklabels([str(v) for v in sweep_vals])
    ax.set_yticks(range(len(benchmarks)))
    ax.set_yticklabels([b.upper() for b in benchmarks])
    ax.set_xlabel(param_label); ax.set_title(title)
    for bi in range(len(benchmarks)):
        for vi in range(len(sweep_vals)):
            if not np.isnan(data[bi, vi]):
                ax.text(vi, bi, f"{data[bi,vi]:.3f}",
                        ha="center", va="center", fontsize=7)
    plt.tight_layout()
    plt.savefig(path, dpi=200, bbox_inches="tight")
    plt.show()
    plt.close()

# ── PART 1 — Main experiment ──────────────────────────────────────────

def run_main(benchmarks):
    print(f"\n{'='*65}")
    print(f"  PART 1  {VARIANT_NAME}  —  Main experiment")
    print(f"  N_RUNS={N_RUNS}  N_PARTICLES={N_PARTICLES}")
    print(f"  ts_size={DEFAULT_TS}  eta={DEFAULT_ETA}  k={DEFAULT_K}")
    print(f"{'='*65}")
    out_dir = f"./results/{VARIANT_NAME}/main"
    os.makedirs(out_dir, exist_ok=True)
    rows = []
    for b_name in benchmarks:
        problem, n_gen, pf, ref_point = get_bench(b_name)
        print(f"\n  [{b_name.upper()}]  n_gen={n_gen}")
        df, best_F = trial_loop(
            problem, n_gen, pf, ref_point,
            run_kwargs=dict(ts_size=DEFAULT_TS, eta=DEFAULT_ETA, k=DEFAULT_K),
            label=f"{VARIANT_NAME}/{b_name.upper()}",
        )
        df.to_csv(f"{out_dir}/{b_name}_metrics.csv", index=False)
        igd_mean, igd_std = df["IGD"].mean(), df["IGD"].std()
        hv_mean,  hv_std  = df["HV"].mean(),  df["HV"].std()
        print(f"    IGD: {igd_mean:.6f} +/- {igd_std:.6f}")
        print(f"    HV:  {hv_mean:.6f} +/- {hv_std:.6f}")
        rows.append({"Function": b_name.upper(),
                     "IGD_Mean": igd_mean, "IGD_Std": igd_std,
                     "HV_Mean":  hv_mean,  "HV_Std":  hv_std})
        plot_pareto(pf, best_F,
                    f"{VARIANT_NAME}  —  {b_name.upper()}  (best of {N_RUNS} runs)",
                    f"{out_dir}/{b_name}_pareto.png")
    summary = save_summary(rows, out_dir, "main")
    print(f"\n{'─'*65}")
    print(f"  Main summary  ({VARIANT_NAME})")
    print(f"{'─'*65}")
    print(summary.to_string(index=False))
    return summary

# ── PART 2 — Sensitivity sweep ────────────────────────────────────────

def _sweep(param_name, sweep_vals, kwarg_key, fixed_kwargs,
           param_label, benchmarks):
    out_dir = f"./results/{VARIANT_NAME}/sensitivity/{param_name}"
    os.makedirs(out_dir, exist_ok=True)
    results, all_rows = {}, []
    for val in sweep_vals:
        print(f"\n  {param_label} = {val}")
        kwargs = {**fixed_kwargs, kwarg_key: val}
        for b_name in benchmarks:
            problem, n_gen, pf, ref_point = get_bench(b_name)
            df, _ = trial_loop(
                problem, n_gen, pf, ref_point, run_kwargs=kwargs,
                label=f"{VARIANT_NAME} {param_label}={val}/{b_name.upper()}")
            df.to_csv(
                f"{out_dir}/{param_name}{val}_{b_name}_metrics.csv", index=False)
            igd_mean, igd_std = df["IGD"].mean(), df["IGD"].std()
            hv_mean,  hv_std  = df["HV"].mean(),  df["HV"].std()
            print(f"    {b_name.upper():6s}  IGD {igd_mean:.5f}+/-{igd_std:.5f}"
                  f"  HV {hv_mean:.5f}+/-{hv_std:.5f}")
            results[(b_name, val)] = {"IGD_Mean": igd_mean, "IGD_Std": igd_std,
                                      "HV_Mean":  hv_mean,  "HV_Std":  hv_std}
            all_rows.append({"Function": b_name.upper(), param_label: val,
                             "IGD_Mean": igd_mean, "IGD_Std": igd_std,
                             "HV_Mean":  hv_mean,  "HV_Std":  hv_std})
    pd.DataFrame(all_rows).to_csv(
        f"{out_dir}/summary_{param_name}.csv", index=False)
    print(f"\n  Best {param_label} per benchmark (lowest mean IGD):")
    for b_name in benchmarks:
        best_val = min(sweep_vals,
            key=lambda v: results.get((b_name,v),{"IGD_Mean":np.inf})["IGD_Mean"])
        print(f"    {b_name.upper():6s}  best {param_label}={best_val}"
              f"  IGD={results[(b_name,best_val)]['IGD_Mean']:.5f}")
    for metric in ("IGD", "HV"):
        plot_line(sweep_vals, results, metric, param_label,
                  f"{VARIANT_NAME}  {param_label} sensitivity  {metric}",
                  f"{out_dir}/line_{metric}.png", benchmarks)
        plot_heatmap(sweep_vals, results, metric, param_label,
                     f"{VARIANT_NAME}  {param_label} heatmap  {metric}",
                     f"{out_dir}/heatmap_{metric}.png", benchmarks)
    print(f"  Saved -> {out_dir}/")

def run_sensitivity(benchmarks):
    print(f"\n{'='*65}")
    print(f"  PART 2  {VARIANT_NAME}  —  Parameter sensitivity")
    print(f"{'='*65}")
    fixed = dict(ts_size=DEFAULT_TS, eta=DEFAULT_ETA, k=DEFAULT_K)
    print("\n  >> Sweeping eta")
    _sweep("eta", ETA_GRID, "eta", fixed, "eta", benchmarks)

# ── Entry point ───────────────────────────────────────────────────────

def run(benchmarks=None):
    if benchmarks is None:
        benchmarks = ALL_BENCHMARKS
    run_main(benchmarks)
    
    # PART 2 is disabled as requested
    # run_sensitivity(benchmarks)
    
    print(f"\n{'#'*65}")
    print(f"  {VARIANT_NAME} complete. Results -> ./results/{VARIANT_NAME}/")
    print(f"{'#'*65}")

if __name__ == "__main__":
    if _in_jupyter:
        run()
    else:
        _args  = [a.lower() for a in sys.argv[1:]]
        _valid = set(ALL_BENCHMARKS)
        _bench = [a for a in _args if a in _valid]
        if _args and not _bench:
            print(f"[ERROR] Unrecognised: {_args}")
            print(f"        Valid: {ALL_BENCHMARKS}")
            sys.exit(1)
        run(benchmarks=_bench if _bench else None)
```
